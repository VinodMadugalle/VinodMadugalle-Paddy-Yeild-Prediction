```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, accuracy_score
import numpy as np

# Assuming df is your DataFrame containing the data

# Features and target
features = ['RainFall', 'Humidity', 'Temperature', 'TotalArea', 'Districts']
target = 'TotalHarvest'

X = df[features]
y = df[target]

# Adding noise to the target variable
np.random.seed(42)
higher_noise = np.random.normal(0, 500, size=len(y))  # Increased noise level
y_higher_noisy = y + higher_noise

# Perform train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_higher_noisy, test_size=0.2, random_state=42)

# Scaling the features if necessary
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the model
model = RandomForestRegressor(n_estimators=5, max_depth=5, random_state=42)
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model performance with increased noise
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error with Higher Noise: {mse}')

# Reducing the threshold for classification (just for example)
threshold = 40000
y_pred_class = (y_pred > threshold).astype(int)
accuracy = accuracy_score(y_test > threshold, y_pred_class)
print(f'Accuracy: {accuracy:.2%}')
```

```python
import numpy as np
# Load new data
months_2024 = [f'2024-{str(month).zfill(2)}' for month in range(1, 13)]
new_data_2024 = pd.DataFrame({
    'YearMonth': months_2024,
    'RainFall': [5, 90, 80, 80, 95, 85, 105, 120, 100, 110, 95, 105],
    'Humidity': [5, 70, 80, 20, 72, 68, 82, 80, 75, 77, 73, 79],
    'Temperature': [5, 25, 30, 25, 30, 28, 29, 31, 25, 29, 27, 28],
    'TotalArea': [20, 180, 210, 150, 210, 200, 215, 230, 210, 225, 205, 240],
    'Districts': [0, 1, 2, 3, 0, 0, 0, 0, 0, 0, 0, 0]
})
# Load the trained model and scaler
model_filename = 'harvest_prediction_model.joblib'
scaler_filename = 'feature_scaler.joblib'
loaded_model = joblib.load(model_filename)
loaded_scaler = joblib.load(scaler_filename)
# Make predictions
new_data_2024['YearMonth'] = pd.to_datetime(new_data_2024['YearMonth'])
new_data_2024_scaled = loaded_scaler.transform(new_data_2024[features])
new_predictions = loaded_model.predict(new_data_2024_scaled)

# Find the month with the highest predicted harvest and its corresponding value
best_month = new_data_2024.loc[new_predictions.argmax(), 'YearMonth']
max_predicted_harvest = new_predictions.max()

print(f'Best month for cultivation based on predicted harvest in 2024 in : ' + district_mapping[new_data_2024.iloc[0, 5]])
print(best_month)
print(f'Maximum Predicted Harvest in 2024: {max_predicted_harvest:.2f}')
```

```python
# Display predictions for all months
new_data_2024['PredictedHarvest'] = new_predictions
new_data_2024['DistrictName'] = new_data_2024['Districts'].map(district_mapping)
print('\nPredicted Harvest for Each Month in 2024:')
print(new_data_2024[['YearMonth', 'PredictedHarvest', 'DistrictName']])
```
